```python
import pandas as pd

df = pd.read_csv(r"C:\Users\lenovo\Downloads\ecommerce_500k.csv")

print("Return Rate (%)")
print(df["is_returned"].value_counts() * 100)
print(df["is_returned"].value_counts(normalize=True) * 100)

print("\nReturn Reasons")
print(df["return_reason"].value_counts())

category_return_rate = pd.crosstab(
    df["category"],
    df["is_returned"],
    normalize="index"
) * 100

print("\nReturn Rate By Category (%)")
print(category_return_rate)
```

```python
print(pd.crosstab(
    df["loyalty_tier"],
    df["is_returned"],
    normalize="index"
) * 100)

city_return_rate = pd.crosstab(
    df["city"],
    df["is_returned"],
    normalize="index"
) * 100

return_col = city_return_rate.columns[-1]
print(city_return_rate.sort_values(by=return_col, ascending=False).head(10))
```
