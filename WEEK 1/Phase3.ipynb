```python
import pandas as pd

csv_path = Path(r"C:\Users\lenovo\Downloads\ecommerce_500k.csv")

if not csv_path.exists():
    raise FileNotFoundError(f"CSV file not found at {csv_path}")

df = pd.read_csv(csv_path)

category_cols = df.select_dtypes(include="string").columns

summary = pd.DataFrame({
    "Column": category_cols,
    "Unique Values": [df[col].nunique() for col in category_cols],
    "Missing Values": [df[col].isnull().sum() for col in category_cols]
})

print(summary.sort_values("Unique Values"))
```
