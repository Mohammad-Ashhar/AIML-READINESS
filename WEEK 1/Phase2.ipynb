```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

candidate_paths = [
    Path.cwd() / "ecommerce_500k.csv",
    Path.cwd() / "WEEK 1" / "ecommerce_500k.csv",
    Path.cwd().parent / "ecommerce_500k.csv",
    Path.cwd().parent / "WEEK 1" / "ecommerce_500k.csv",
]

data_path = None
for path in candidate_paths:
    if path.exists():
        data_path = path
        break

if data_path is None:
    raise FileNotFoundError("CSV file not found. Place ecommerce_500k.csv in the notebook folder or update the path.")

df = pd.read_csv(data_path)

print("Skewness")
print(df["final_price"].skew())

plt.hist(df["final_price"])
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```
