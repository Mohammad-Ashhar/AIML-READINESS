# E-commerce Dataset Analysis

This notebook is GitHub-friendly and uses a portable relative path to locate the CSV file.

```python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path.cwd() / "ecommerce_500k.csv",
    Path.cwd() / "WEEK 1" / "ecommerce_500k.csv",
    Path.cwd().parent / "ecommerce_500k.csv",
    Path.cwd().parent / "WEEK 1" / "ecommerce_500k.csv",
]

data_path = None
for path in candidate_paths:
    if path.exists():
        data_path = path
        break

if data_path is None:
    raise FileNotFoundError("CSV file not found. Place ecommerce_500k.csv in the notebook folder or update the path.")

df = pd.read_csv(data_path)

print(df.shape)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print("Duplicates:")
print(df.duplicated().sum())
```
